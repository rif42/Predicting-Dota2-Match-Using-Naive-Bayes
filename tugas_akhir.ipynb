## Prediksi hasil match berdasarkan hero yang di pilih di Dota 2

Dota adalah sebuah game ber genre MOBA dengan platform PC yang sangat populer.  
Di sebuah game Dota, terdapat 2 tim yang berlawanan, yaitu Radiant dan Dire. Masing masing team mempunyai 5 buah player.
Dota mempunyai banyak hal yang dapat mempengaruhi hasil match, seperti skill, strategi, item, timing dan sebagainya.

Namun di tugas ini, kita hanya akan membahas hero yang di pilih oleh satu sisi.
Dota 2 mempunyai total hero sebanyak 123 hero.
Satu tim akan memilih 5 hero untuk dimainkan, dan kita akan menghitung presentase kemenangan tim tersebut berdasarkan hero yang di pilih.

Kita mempunyai data 10.000 match dari Dota 2 yang berlangsung di seluruh penjuru dunia.
Match tersebut tersebar dari beberapa patch yang berbeda, dan juga berbeda tahun.
Di setiap patch, ada hero baru, item baru, dan juga nerf hero atau item.
Perbedaan antar patch ini akan mempengaruhi prediksi algoritma kita.

Data ini diambil langsung dari Dota 2 API.

```python
import pandas as pd

df = pd.read_csv('global.csv')
df.head()
```

### Step 1 : Pilih kolom yang dibutuhkan

```python
df1 = pd.DataFrame(df, columns=["Match ID", "Team A Heroes", "Winner"])
df1.head()
```

### Step 2 : Rename kolom

```python
#rename columns

df2 = pd.DataFrame(df1)
df2.rename(columns={'Match ID': 'match_id', 'Team A Heroes': 'hero1', 'Winner': 'result'}, inplace=True)
df2.head()
```

### Step 3 : Buat kolom baru untuk 5 hero di satu tim

```python
df3 = pd.DataFrame(df2)
df3.insert(2, 'hero2', 0)
df3.insert(3, 'hero3', 0)
df3.insert(4, 'hero4', 0)
df3.insert(5, 'hero5', 0)

df3.head()
```

### Step 4 : Proses data lebih lanjut

- Rename data di kolom result menjadi 1 dan 0. 1 berarti menang, 0 berarti kalah
- Split data di hero1, lalu sort berdasarkan alphabet
- hapus space yang ada di depan dan belakang nama hero
- lalu masukkan ke dalam kolom hero 2-5

```python
#loop for each row, adding heroes to new columns

df4 = pd.DataFrame(df3)

#change result column to 1 and 0. 1 means winning, 0 means losing
for i in range(len(df4)): 
    if (df4.loc[i,'result'] == "Team A"):
        df4.loc[i,'result'] = 1
    elif (df4.loc[i,'result'] == "Team B"):
        df4.loc[i,'result'] = 0
        
    data=df4.loc[i,'hero1']
    data=data.split("  ")
    data.sort()
    
    df4.loc[i,'hero1']=data[0].strip()
    df4.loc[i,'hero2']=data[1].strip()
    df4.loc[i,'hero3']=data[2].strip()
    df4.loc[i,'hero4']=data[3].strip()
    df4.loc[i,'hero5']=data[4].strip()
    
df4.head()
```

### Step 5 : Buat ID untuk hero yang unik

- Kita akan mengambil semua hero yang ada di kolom hero1-5, dari semua match.
- Lalu kita hapus duplikat yang ada di dalam data.
- Selanjutnya kita masukkan semua data string ke dalam sebuah list, dan urutkan berdasarkan alphabet

hasilnya hero yang mempunyai awalan A akan berada di index/id bagian awal
Sebagai contoh :  
Abaddon mempunyai id 0
Alchemist mempunyai id 1  
Bounty Hunter mempunyai id 10


```python
#make unique ID for each hero
df5 = pd.DataFrame(df4, columns=["hero1", "hero2", "hero3", "hero4", "hero5"])

allheroes=[]
for i in range(len(df5)):
    allheroes.append(df5.loc[i,'hero1'])
    allheroes.append(df5.loc[i,'hero2'])
    allheroes.append(df5.loc[i,'hero3'])
    allheroes.append(df5.loc[i,'hero4'])
    allheroes.append(df5.loc[i,'hero5'])
    
allheroes = list(dict.fromkeys(allheroes))
allheroes = sorted(allheroes)
print(allheroes)
print(len(allheroes))
```

### Step 6 : Ubah data nama hero menjadi angka (id)

- Loop semua data di kolom hero1-5
- jika nama hero ada di list allheroes, ambil index/id dari hero tersebut
- gantikan nama hero dengan index/id hero tersebut

```python
#replace hero names with hero id

df6 = pd.DataFrame(df4)

for i in range(len(df6)):
    x=(df6.loc[i,'hero1'])
    xx=allheroes.index(x, 0, len(allheroes))
    df6.loc[i,'hero1'] = xx
    
    y=(df6.loc[i,'hero2'])
    yy=allheroes.index(y, 0, len(allheroes))
    df6.loc[i,'hero2'] = yy
    
    z=(df6.loc[i,'hero3'])
    zz=allheroes.index(z, 0, len(allheroes))
    df6.loc[i,'hero3'] = zz
    
    c=(df6.loc[i,'hero4'])
    cc=allheroes.index(c, 0, len(allheroes))
    df6.loc[i,'hero4'] = cc
    
    d=(df6.loc[i,'hero5'])
    dd=allheroes.index(d, 0, len(allheroes))
    df6.loc[i,'hero5'] = dd

df6.head()
```

```python
df6.to_csv(r'./finaldata.csv', index=False)
```

### Step 7 : load ke pandas, hilangkan kolom matchid

```python
import pandas as pd

data = pd.read_csv('./finaldata.csv')
data = data.drop(['match_id'], axis=1)
data.to_csv(r'./finaldata.csv', index=False)

data.head()
```

### Step 8 : Split data menjadi data train dan data test

```python
from sklearn.model_selection import train_test_split

X = data.drop('result',axis=1)
y = data["result"]
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.1, random_state = 1)
```

### Step 9 : buat model berdasarkan algoritma naive bayes

**Naive Bayes Gaussian**  
 Algoritma ini mengasumsikan bahwa data mengikuti distribusi normal dan menghitung kemungkinan setiap kelas berdasarkan rata-rata dan standar deviasi fitur untuk kelas tersebut.

**Naive Bayes Multinomial**  
 Algoritma ini digunakan untuk tugas klasifikasi di mana fitur-fiturnya diskrit dan multinomial (misalnya, hitungan kata dalam dokumen). Algoritma ini menghitung kemungkinan setiap kelas berdasarkan frekuensi setiap fitur dalam data latihan.

**Naive Bayes Bernoulli**  
 Algoritma ini mirip dengan algoritma naive Bayes multinomial, tetapi digunakan untuk tugas klasifikasi di mana fitur-fiturnya biner (yaitu, hanya memiliki dua nilai). Algoritma ini menghitung kemungkinan setiap kelas berdasarkan adanya atau tidaknya setiap fitur dalam data latihan.

 **Naive Bayes Complementary**  
 adalah algoritma yang didasarkan pada prinsip yang sama seperti naive Bayes, tetapi berbeda dalam beberapa cara khusus. Sebagai contoh, mungkin ada varian dari algoritma naive Bayes yang menggunakan distribusi yang berbeda untuk memodelkan data (misalnya, distribusi gamma bukan distribusi normal) atau yang melepaskan asumsi independensi antar fitur.

```python
from sklearn.naive_bayes import GaussianNB, BernoulliNB, MultinomialNB, ComplementNB
from sklearn.metrics import accuracy_score
import pickle5 as pickle

nb = GaussianNB()
nb.fit(X_train, y_train)
nb_predict = nb.predict(X_test)

nb_b = BernoulliNB()
nb_b.fit(X_train, y_train)
nb_b_predict = nb_b.predict(X_test)

nb_m = MultinomialNB()
nb_m.fit(X_train, y_train)
nb_m_predict = nb_m.predict(X_test)

nb_c = ComplementNB()
nb_c.fit(X_train, y_train)
nb_c_predict = nb_c.predict(X_test)

nb_acc_score = accuracy_score(y_test, nb_predict)
nb_b_score = accuracy_score(y_test, nb_b_predict)
nb_m_score = accuracy_score(y_test, nb_m_predict)
nb_c_score = accuracy_score(y_test, nb_c_predict)

with open('nb_gaussian.pkl', 'wb') as file:
  pickle.dump(nb, file)
  
with open('nb_bernoulli.pkl', 'wb') as file:
  pickle.dump(nb_b, file)
  
with open('nb_multinomial.pkl', 'wb') as file:
  pickle.dump(nb_m, file)
  
with open('nb_complementary.pkl', 'wb') as file:
  pickle.dump(nb_c, file)

print(f'naive bayes : {nb_acc_score}')
print(f'bernoulli : {nb_b_score}')
print(f'multinomial : {nb_m_score}')
print(f'complementary : {nb_c_score}')
```

### Step 10 : Buat model berdasarkan algoritma SVM

**Apa itu SVM?**

Support vector machines (SVMs) adalah sebuah algoritma yang digunakan untuk klasifikasi dan regresi. SVMs adalah salah satu algoritma yang paling populer untuk machine learning, karena algoritma ini sangat fleksibel dan dapat digunakan untuk klasifikasi linier dan non-linier, regresi, dan deteksi outlier.

Parameter yang digunakan adalah :

- Kernel, adalah parameter yang paling penting dari SVM. Kernel adalah fungsi yang digunakan untuk menghitung jarak antara dua data. Kernel yang digunakan adalah linear, polynomial, dan RBF.

- regularization (C), adalah parameter yang digunakan untuk mengatur trade-off antara memperbolehkan model untuk menyesuaikan data latih dengan baik dan mencegah overfitting. Semakin besar nilai C, maka model akan mencoba untuk menyesuaikan data latih dengan baik, yang dapat menyebabkan overfitting. Semakin kecil nilai C, maka model akan lebih sensitif terhadap noise di data latih, yang dapat meningkatkan generalisasi ke data baru.

- gamma/degree, adalah parameter spesifik untuk model polynomial dan RBF yang digunakan untuk mengontrol lebar dari filter gauss. Semakin besar nilai gamma, maka model akan lebih sensitif terhadap perubahan data latih, yang dapat menyebabkan overfitting. Semakin kecil nilai gamma, maka model akan lebih sensitif terhadap noise di data latih, yang dapat meningkatkan generalisasi ke data baru.
